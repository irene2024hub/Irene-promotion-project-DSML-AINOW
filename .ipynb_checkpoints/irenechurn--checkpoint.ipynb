

# Age vs Churn Insights

- **Most Customers Are Young**: The majority of customers fall between 25 and 50 years old, with peaks around 30 and 40.
- Churn Increases with Age: **The green bars (churned customers) become more prominent as age increases**, especially
-  beyond 50 years.
- Younger Customers Stay: **Orange bars (non-churned) dominate in the 2540 age range**, indicating younger customers are more likely to stay.
- Older Customers Are at Risk: Churn rates rise steadily with age, suggesting older customers are more likely to leave the bank.

 Strategic Implications
- Retention Focus: Target retention efforts on customers aged 50 and above, who show higher churn risk.
- Engagement Programs: Younger customers are more stableconsider loyalty programs to maintain their satisfaction.
- Age-Sensitive Services: Tailor services and communication styles to different age groups to reduce churn.


```python
# Create age group column
import pandas as pd

# Define age bins and labels
bins = [20, 40, 60, 70, 90]
labels = ['20-40', '41-60', '61-70', '71-90']

# Create age group column
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Group by AgeGroup and Churn status
Churn_stats = df.groupby(['AgeGroup', 'Churn']).size().unstack(fill_value=0)

# Add total and churn rate
Churn_stats['Total'] = churn_stats.sum(axis=1)
Churn_stats['ChurnRate (%)'] = (Churn_stats[1] / Churn_stats['Total'] * 100).round(2)

print(Churn_stats)
```

```python
# Group by AgeGroup and Churn status
Age_Churn = df.groupby(['AgeGroup', 'Churn']).size().reset_index(name='Count')

# Plot
plt.figure(figsize=(8, 4))
sns.barplot(data=Age_Churn, x='AgeGroup', y='Count', hue='Churn', palette='Set2')
plt.title('Churn Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Customers')

# Get handles and labels, then relabel with colors preserved
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['No (0)', 'Yes (1)'], title='Churn')

plt.tight_layout()
plt.show()
```

# CORRELATIONS 
**Correlation** is a statistical measure that describes the strength and direction of a relationship between two variables.

If two variables tend to move together, they are positively correlated. If one increases while the other decreases, they are negatively correlated. If there's no consistent pattern, they are uncorrelated.

 Key Points:
- Range: Correlation values range from -1 to +1:
1. +1: Perfect positive correlation (both increase together)
2. 0: No correlation (no relationship)
3 -1: Perfect negative correlation (one increases, the other decreases)
   
 Does NOT imply **causation**: Just because two variables are correlated doesnt mean one causes the other.

 Based on the correlation heatmap from the  churn modeling dataset, here are the features that show 
 **notable correlation with the target variable (Churn)**:

###  Features Positively Correlated with Churn

- **Balance**: **0.12**  Higher balances are slightly associated with higher churn.
- **Gender_Male**: **0.11**  Males are slightly more likely to churn.

###  Features Negatively Correlated with Churn
These features decrease the likelihood of churn:
- **NumOfProducts**: **-0.04**  Customers with more products are less likely to churn.
- **Geography_Spain**: **-0.053**  Spanish customers are less likely to churn.

###  Features with Weak or No Correlation

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv("Churn_Modelling.csv")

# Select relevant numerical features
features_to_scale = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
X_raw = df[features_to_scale]
y = df['Churn']

# Apply StandardScaler
scaler = StandardScaler()
scaled_features = scaler.fit_transform(X_raw)
X_scaled = pd.DataFrame(scaled_features, columns=features_to_scale)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

```

```python
# check training dataeset
print("Training set:", X_train.shape, y_train.shape)
```

```python
# check testing dataset

print("Testing set:", X_test.shape, y_test.shape)
```

STEP 5

# MODEL TRAINING AND EVALUATION USING: 

```python
# 1. Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix



```

```python
# 6. Train Logistic Regression
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
log_preds = log_model.predict(X_test)
```

```python
print(" Logistic Regression Accuracy:", accuracy_score(y_test, log_preds))
```

```python
# 7. Train Decision Tree
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
tree_preds = tree_model.predict(X_test)
```

```python
print(" Decision Tree Accuracy:", accuracy_score(y_test, tree_preds))
```

```python
# 8. Train Random Forest
forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
forest_model.fit(X_train, y_train)
forest_preds = forest_model.predict(X_test)
```

```python
print(" Random Forest Accuracy:", accuracy_score(y_test, forest_preds))
```

```python
# Optional: Detailed classification report
print("\n Classification Report (Random Forest):")
print(classification_report(y_test, forest_preds))
```

```python
# Optional: Confusion Matrix
print("\n Confusion Matrix (Random Forest):")
print(confusion_matrix(y_test, forest_preds))
```

```python
# Import XGBOOST 
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2. Initialize and train the model
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train, y_train)

# 3. Make predictions
xgb_preds = xgb_model.predict(X_test)

# 4. Evaluate performance
print(" XGBoost Accuracy:", accuracy_score(y_test, xgb_preds))
print("\n Classification Report:")
print(classification_report(y_test, xgb_preds))
print("\n Confusion Matrix:")
print(confusion_matrix(y_test, xgb_preds))
```
